```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score 
```

```python
df = pd.read_csv('Movie_classification.csv')
df.head()
```

```python
df.info()
```

De la tabla anterior, se puede apreciar que la variable **"Time_takes"** contiene datos faltantes, y existen variables categóricas; **"3D_available"**, y **"Genre"**.

# **Tratamiento de datos faltantes**

Dado que los valores faltantes son numéricos, vamos a emplear el método **.fillna()** para sustituir los valores faltantes por la media correspondiente a cada una de las variables.

```python
mean = df['Time_taken'].mean()
mean
```

```python
# imputamos datos faltantes por la media
df['Time_taken'] = df['Time_taken'].fillna(mean) 
```

```python
# verificamos que ya no existan datos faltantes
df.info()
```

Debido a que no existen datos faltantes, procedemos a transformar variables categóricas a numéricas.

# **Variables categóricas**

```python
df.head()
```

```python
df =  pd.get_dummies(df, ['3D_available', 'Genre'], drop_first=True)
df.head()
```

```python
freq_class = df.groupby('Start_Tech_Oscar')['Start_Tech_Oscar'].count()
freq_class
```

```python
plt.figure(figsize=(12,6))
plt.bar(freq_class.index, freq_class)
plt.show()
```

# **Separación X, y**

```python
X = df.drop('Start_Tech_Oscar', axis=1)
y = df['Start_Tech_Oscar']
```

# **División train_test**

```python
# separación en conjunto de entrenamiento y conjunto de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# **XGBoosting Classifier (clasificación)**

Parámetros:

- **learning_rate** : La tasa de aprendizaje $learning\_rate$ reduce la contribución de cada árbol. Hay un equilibrio entre tasa de aprendizaje y $n\_estimadores$. Los valores deben estar en el rango (0.0, inf)

- **max_depth** : Determina la porfundidad permitida por el árbol durante el inicio de cualquier boosting.

- **subsample** : Porcentaje de la muestra usada por árbol un valor bajo puede llevar a un subajuste.

- **colsample_bytree** : Porcentaje de características empleadas por árbol. Un valor alto puede llevar a un sobreajuste.

- **n_estimators** : Número de árboles a generar.

- **objective** : Determina la función de pérdida empleada. 


**Referencia:** [sklearn.ensemble.AdaBoostClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.AdaBoostClassifier.html)

```python
xgb = XGBClassifier(n_estimators=100, learning_rate=0.02)
xgb.fit(X_train, y_train)
```

# **Predicción**

```python
y_pred_train = xgb.predict(X_train)
y_pred_test = xgb.predict(X_test)
```

# **Rendimiento (Performance)**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
  # metrics for train
accuracy = accuracy_score(y_train, y_pred_train)
precision = precision_score(y_train, y_pred_train)
recall = recall_score(y_train, y_pred_train)
f1score = f1_score(y_train, y_pred_train)
confusionm = confusion_matrix(y_train, y_pred_train)

print('Accuracy: {}'.format(accuracy))
print('Precision: {}'.format(precision))
print('Recall: {}'.format(recall))
print('F1score: {}'.format(f1score))
```

```python
confusionm
```

```python
# Confusion matrix for Random Forest
target_labels = ['0','1']

plt.figure(figsize=(10,6))
cm = confusion_matrix(y_train, y_pred_train)
s_cm = sns.heatmap(cm,cmap='plasma',annot=True, fmt='g',
            xticklabels=target_labels,
            yticklabels=target_labels)
s_cm.set(xlabel='Predicted',ylabel='Real')
plt.show()
```

```python
# metrics for test
accuracy = accuracy_score(y_test, y_pred_test)
precision = precision_score(y_test, y_pred_test)
recall = recall_score(y_test, y_pred_test)
f1score = f1_score(y_test, y_pred_test)
confusionm = confusion_matrix(y_test, y_pred_test)

print('Accuracy: {}'.format(accuracy))
print('Precision: {}'.format(precision))
print('Recall: {}'.format(recall))
print('F1score: {}'.format(f1score))
```
